# Perfect Correlation Among Price Columns:
The columns Open, High, Low, Close, and Adj Close all have a perfect correlation of 1 with each other. This indicates a perfect linear relationship among these price-related columns.

# Volume Correlation:
The Volume column shows a very low correlation (around 0.08) with all the price-related columns (Open, High, Low, Close, Adj Close). This low correlation indicates that the trading volume does not have a strong linear relationship with the stock prices.

## Analysis Using 44 and 100 Days Moving Average

```python
Tesla_adj_close_ma =Tesla['Adj Close']
Tesla['44'] = Tesla_adj_close_ma.rolling(window=44).mean()
Tesla['100'] = Tesla_adj_close_ma.rolling(window=100).mean()
# Plot the adjusted closing price, the short and long windows of rolling means
Tesla[['Adj Close', '44', '100']].plot(title="Tesla")
plt.title('Tesla')
plt.ylabel('Adjusted Closing Price')
plt.show()
```
